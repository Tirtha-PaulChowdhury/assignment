#### 1) We have the min and max temperatures in a city In India for each months of the year. We would like to find a function to describe this and show it graphically, the dataset given below.
#### Task:
##### 1.fitting it to the periodic function
##### 2.plot the fit
#### Data
##### Max = 39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25
##### Min = 21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

temp_max = np.array([39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25])
temp_min = np.array([21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18])
month = np.arange(12)

plt.plot(month , temp_max , 'ro')
plt.plot(month , temp_min , 'co')
plt.xlabel('Months')
plt.ylabel('Min and Max Temps')
plt.show()
```

```python
# 1) Fitting the plot to a pereodic function

from scipy import optimize
def yearly_temps(times, avg, ampl, time_offset):
    return (avg+ampl*np.cos((times + time_offset) * 1.8 * np.pi / times.max()))

res_max , cov_max = optimize.curve_fit(yearly_temps , month , temp_max , [40 , 20 , 0])
res_min , cov_min = optimize.curve_fit(yearly_temps , month , temp_min , [-40 , 20 , 0])
```

```python
# 2) Plotting the fit

days = np.linspace(0,12,365)

plt.figure()
plt.plot(month , temp_max , 'ro')
plt.plot(month , temp_min , 'co')
plt.plot(days , yearly_temps(days, *res_max) , '-r')
plt.plot(days , yearly_temps(days, *res_min) , '-c')
plt.xlabel('Months')
plt.ylabel('Min and Max Temps')
plt.show()
```

#### 2) Charts to plot:
##### 1. Create a pie chart presenting the male/female proportion
##### 2. Create a scatterplot with the Fare paid and the Age, differ the plot color by gender

```python
import pandas as pd
titanic = pd.read_csv('https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv')
titanic.head()
```

```python
# 1) pie chart plotting

count = titanic['sex'].value_counts()
colors = ['orange' , 'grey']


plt.pie(count , labels = count.index.values.tolist(), colors = colors , autopct = '%.2f%%' )
plt.title('Male/Female proportions')
```

```python
#converting male = 0 and female = 1 as we need numerical values.

titanic.replace(to_replace = 'male' , value = 0 , inplace = True)
titanic.replace(to_replace = 'female' , value = 1 , inplace = True)
```

```python
# 2) Scatterplot

plt.scatter(titanic['fare'] , titanic['age'] , c=titanic['sex'] , cmap = 'copper' , alpha = 0.6)
plt.colorbar().set_label('Male = 0 , Female = 1')
plt.title('Scatterplot')
plt.xlabel('Fare')
plt.ylabel('Age')
```
